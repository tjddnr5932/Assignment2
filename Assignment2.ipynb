```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('C:/Users//Downloads/Region.csv')
new_df = df.iloc[:243]
pro_ele = new_df.groupby('province')['elementary_school_count'].sum()
pro_kin = new_df.groupby('province')['kindergarten_count'].sum()
pro_uni = new_df.groupby('province')['university_count'].sum()

fig = plt.figure()
ax = plt.subplot(1, 1, 1)
ax.set_title('province/elementray count')
for label in ax.xaxis.get_ticklabels() :
    label.set_rotation(90)
ax.plot(pro_ele)

fig = plt.figure()
ax = plt.subplot(1, 1, 1)
ax.set_title('province/kindergarten count')
for label in ax.xaxis.get_ticklabels() :
    label.set_rotation(90)
ax.plot(pro_kin)

fig = plt.figure()
ax = plt.subplot(1, 1, 1)
ax.set_title('province/university count')
for label in ax.xaxis.get_ticklabels() :
    label.set_rotation(90)
ax.plot(pro_uni)


scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(new_df['elderly_population_ratio'],new_df['elderly_alone_ratio'])
axes1.set_title('Scatterplot of elderly_population_ratio vs elderly_alone_ratio')
axes1.set_xlabel('elderly_population_ratio')
axes1.set_ylabel('elderly_alone_ratio')
```
